## # GOLD DIMENSION: PRODUCTS

This dimension provides a unified, analytics-ready view of products by combining
product information from CRM and ERP systems.


```python
%sql
-- Preview source Silver table for Products dimension
SELECT *
FROM analytics.silver.crm_products
LIMIT 10;
```

```python
%sql
-- Preview ERP category data used for product enrichment
SELECT *
FROM analytics.silver.erp_category
LIMIT 10;
```

## -- Create Gold Products Dimension

```python
%sql
-- Create Gold Products Dimension
-- Analytics-ready product dimension based on CRM data
-- One row per product

CREATE OR REPLACE TABLE dim_products AS
SELECT
    product_id,
    product_key,
    product_name,
    product_line,
    product_cost
FROM analytics.silver.crm_products;
```

## Previw Data

```python
%sql
SELECT *
FROM dim_products
LIMIT 10;
```
