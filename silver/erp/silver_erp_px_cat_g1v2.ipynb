```python
%sql
USE CATALOG analytics;
USE SCHEMA silver;
```

## Read Bronze Table

```python
df = spark.table("analytics.bronze.erp_px_cat_g1v2")
df.display()
```

## Trim all string columns

```python
from pyspark.sql.functions import trim, col
from pyspark.sql.types import StringType

for f in df.schema.fields:
    if isinstance(f.dataType, StringType):
        df = df.withColumn(f.name, trim(col(f.name)))

df.display()
```

## Rename Columns

```python
df = (
    df
    .withColumnRenamed("CID", "category_id")
    .withColumnRenamed("CAT", "category")
    .withColumnRenamed("SUBCAT", "subcategory")
    .withColumnRenamed("SUBCAT", "subcategory")
    .withColumnRenamed("MAINTENANCE", "maintenance")
)

df.display()
```

## Write Silver Table

```python
df.write \
  .mode("overwrite") \
  .format("delta") \
  .saveAsTable("analytics.silver.erp_category")
```
